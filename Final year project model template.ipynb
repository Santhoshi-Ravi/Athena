# Features

We load the file path name as well as the actual phoneme from the timit database

```python
data_label = pd.read_csv("audio scaled timit.csv")
```

We only consider one audio sample for making our predictions, that audio sample alone is presented here

```python
data_label[6160:]
```

```python
actual_phoneme = data_label['0'][6160:]
#this is our actual phonemes given in the database
```

Extracting features from the given audio sample; we are slicing the audio sample based on time inputs so that we get audio sample of each phoneme 

```python
#convert the test audio sample here we consider dr8-fbcg1/si2242 as our sample audio
m=[]
c=[]
mels=[]
con=[]
ton=[]
for i in range(6160,6185):
    path ='G:/timit/timit/'+data_label['file name'][i]+'.wav'
    X,sr = librosa.load(path)
    sample_rate = sr
    n_fft = 64
    start = data_label['start time scale'][i]
    end = data_label['end time scale'][i]
    mfccs = np.mean(librosa.feature.mfcc(y=X[start:end], sr=sample_rate, n_mfcc=26, n_fft =64,n_mels=26).T,axis=0)
# Generates a Short-time Fourier transform (STFT) to use in the chroma_stft
    stft = np.abs(librosa.stft(X))
# Computes a chromagram from a waveform or power spectrogram.
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
# Computes a mel-scaled spectrogram.
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
# Computes spectral contrast
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
# Computes the tonal centroid features (tonnetz)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X),
    sr=sample_rate).T,axis=0)
    m.append(mfccs)
    c.append(chroma)
    mels.append(mel)
    con.append(contrast)
    ton.append(tonnetz)
    print(i,mfccs.shape,start,end)
```

```python
features = []
for i in range(len(ton)):
    features.append(np.concatenate((m[i], c[i], ton[i], con[i], mels[i]), axis=0))
    print(i)
```

```python
X = features
```

# Model 

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X = ss.fit_transform(X)
```

We have already created a model, we upload that model to make predictions of the phoneme pronounced

```python
from tensorflow.keras.models import model_from_json
```

```python
# load json and create model
json_file = open('model_13.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model_13.h5")
print("Loaded model from disk")
 
```

```python
# evaluate loaded model on test data
loaded_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
predictions = loaded_model.predict_classes(X)
```
